```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
acc_knn_rnd_forest = pd.read_excel("accuracy_knn_rand_forest.xlsx")
acc_knn_rnd_forest

acc_knn_9 = acc_knn_rnd_forest[9][0]
acc_knn_12 = acc_knn_rnd_forest[12][0]
acc_knn_15 = acc_knn_rnd_forest[15][0]
acc_knn_18 = acc_knn_rnd_forest[18][0]


acc_rnd_forest_9  = acc_knn_rnd_forest[9][1]
acc_rnd_forest_12 = acc_knn_rnd_forest[12][1]
acc_rnd_forest_15 = acc_knn_rnd_forest[15][1]
acc_rnd_forest_18 = acc_knn_rnd_forest[18][1]

acc_rnd_forest = [acc_rnd_forest_9, acc_rnd_forest_12, acc_rnd_forest_15, acc_rnd_forest_18]
acc_knn = [acc_knn_9, acc_knn_12, acc_knn_15, acc_knn_18]

acc_knn
```

```python
acc_nn_9 = np.load("metrics_dim_9.npy", allow_pickle=True).item()['test_accuracy']
acc_nn_12 = np.load("metrics_dim_12.npy", allow_pickle=True).item()['test_accuracy']
acc_nn_15 = np.load("metrics_dim_15.npy", allow_pickle=True).item()['test_accuracy']
acc_nn_18 = np.load("metrics_dim_18.npy", allow_pickle=True).item()['test_accuracy']

acc_nn = [acc_nn_9, acc_nn_12, acc_nn_15, acc_nn_18]
acc_nn
```

```python
# Data from the user's description
n_values = [9, 12, 15, 18]
target_accuracies = [0.95, 0.9, 0.8, 0.7]

# Plotting
plt.figure(figsize=(8, 5))

# Plotting Target Accuracies
plt.plot(n_values, target_accuracies, marker='o', color='red', linestyle='-', label='Target Accuracies')
plt.scatter(n_values, target_accuracies, color='red', s=50)

# Plotting KNN Accuracies
plt.plot(n_values, acc_knn, marker='s', color='blue', linestyle='--', label='K-Nearest Neighbors')
plt.scatter(n_values, acc_knn, color='blue', s=50)

# Plotting Random Forest Accuracies
plt.plot(n_values, acc_rnd_forest, marker='^', color='green', linestyle='-.', label='Random Forest')
plt.scatter(n_values, acc_rnd_forest, color='green', s=50)

# Plotting Neural Network Accuracies
plt.plot(n_values, acc_nn, marker='d', color='purple', linestyle=':', label='Neural Network')
plt.scatter(n_values, acc_nn, color='purple', s=50)

# Adding annotations for each data point
for i, n in enumerate(n_values):
    plt.text(n, target_accuracies[i] + 0.01, f'{target_accuracies[i]:.3f}', color='red', ha='center')
    plt.text(n, acc_knn[i] + 0.01, f'{acc_knn[i]:.3f}', color='blue', ha='center')
    plt.text(n, acc_rnd_forest[i] + 0.01, f'{acc_rnd_forest[i]:.3f}', color='green', ha='center')
    plt.text(n, acc_nn[i] + 0.01, f'{acc_nn[i]:.3f}', color='purple', ha='center')

# Labels and title
plt.xlabel("Feature Dimension of Kryptonite-n (n)")
plt.ylabel("Accuracy")
plt.title("Comparison of Model Accuracies")
plt.ylim(0.5, 1.0)
plt.grid(True)
plt.legend()
plt.show()
```
