```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_gaussian_columns(data, feature_names=None):
    """
    Plots Gaussian distributions for each column in the dataset.
    
    Parameters:
    - data: numpy array, each column is a feature.
    - feature_names: list of strings, optional names for each column.
    """
    num_columns = data.shape[1]

    if feature_names is None:
        feature_names = [f'Feature {i+1}' for i in range(num_columns)]
        
    for i in range(num_columns):
        col_data = data[:, i]
        
        # Plot the distribution
        plt.figure(figsize=(8, 4))
        sns.histplot(col_data, kde=False, stat='density', bins=30, color='skyblue', label='Data')
        
        plt.title(f'Distribution for {feature_names[i]}', fontsize=22)
        plt.xlabel(feature_names[i], fontsize=22)
        plt.ylabel('Density', fontsize=22)
        plt.legend()
        plt.show()
```

```python
X = np.load(fr'../../Datasets/kryptonite-{9}-X.npy')

plot_gaussian_columns(X)
```

```python
def plot_gaussian_classes_together(class_1_data, class_2_data, feature_names=None):
    """
    Plots Gaussian distributions for each feature column, overlaying both classes.
    
    Parameters:
    - class_1_data: numpy array, data for class 1 (e.g. y == 0).
    - class_2_data: numpy array, data for class 2 (e.g. y == 1).
    - feature_names: list of strings, optional names for each column.
    """
    num_columns = class_1_data.shape[1]

    if feature_names is None:
        feature_names = [f'Feature {i+1}' for i in range(num_columns)]
        
    for i in range(num_columns):
        col_data_1 = class_1_data[:, i]
        col_data_2 = class_2_data[:, i]
        
        plt.figure(figsize=(8, 4))
        sns.histplot(col_data_1, kde=False, stat='density', bins=30, color='blue', alpha=0.5, label='Class 1')
        sns.histplot(col_data_2, kde=False, stat='density', bins=30, color='red', alpha=0.5, label='Class 2')
        
        plt.title(f'Distributions for {feature_names[i]} (Class 0 vs Class 1)', fontsize=22)
        plt.xlabel(feature_names[i], fontsize=22)
        plt.ylabel('Density', fontsize=22)
        plt.legend(bbox_to_anchor=(1.15, 0), fontsize=22)
        plt.show()
```

```python
y = np.load(f'../../Datasets/kryptonite-{9}-y.npy')
X = np.load(f'../../Datasets/kryptonite-{9}-X.npy')

class_1_data = X[y == 0, :]
class_2_data = X[y == 1, :]

class_1_data.shape
```

```python
plot_gaussian_classes_together(class_1_data, class_2_data)
```
