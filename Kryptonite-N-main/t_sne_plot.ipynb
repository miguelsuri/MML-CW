```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load dataset
n = 15
X = np.load(f'Datasets/kryptonite-{n}-X.npy')
y = np.load(f'Datasets/kryptonite-{n}-y.npy')

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Perform t-SNE with 2 components for visualization
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, max_iter=250, random_state=42, n_jobs=-1)
X_tsne = tsne.fit_transform(X)

# Plot the 2D projection of the data
plt.figure(figsize=(10, 7))
for label in np.unique(y):
    plt.scatter(X_tsne[y == label, 0], X_tsne[y == label, 1], label=f'Class {label}', alpha=0.1)
    
plt.title("2D Projection of Data using t-SNE")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()
plt.show()
```
